Finding Factorial

```python
n=int(input("Enter the no  :"))
fact=1
i=1
while i<=n:
    fact=fact*i
    i=i+1
print(fact)
    
```

```python
number=int(input("Enter the no"))
a,b=0,1
counter=0
while counter< number:
    print(a)
    c=a+b
    a=b
    b=c
    counter=counter+1
    
```

```python
n=int(input("Enter the No :"))
a,b=0,1
for i in range (n):
    print(a)
    c=a+b
    a=b
    b=c
```
