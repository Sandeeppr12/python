```python
##Multiple Inheritance >> one child class inherit from multipal parent class
```

```python
class Parent_class1:
    def parent_class1(self):
        print("parent class 1 info")
class Parent_class2:
    def parent_class2(self):
        print("parent class 2 info")
class Parent_class3:
    def parent_class3(self):
        print("parent class 3 info")
```

```python
class Child_class(Parent_class1,Parent_class2,Parent_class3):
    def ChildMethod(self):
        print("This is child info")
```

```python
child=Child_class()
```

```python
child.ChildMethod()
```

```python
child.parent_class1()
```
