```python
def test_feb(n):
    a,b=0,1
    for i in range(n):
        yield a
        a,b=b,a+b
```

```python
fib=test_feb(10)
```

```python
for i in test_feb(10):
    print(i)
```

```python
for i in range(10):
    print(next(fib) ,end=" ")
```
