```python
class calculator:
    def sum(self,a,b):
        return a+b
    def mul(self,a,b):
        return a*b
    def diff(self,a,b):
        if a>b:
            return a-b
        else:
            return b-a
```

```python
class teacher_lec:
    def lec_info(self):
        print("this is lec info with teacher perspective")
        
```

```python
class Student_lec:
    def lec_info(self):
        print("this is lec info with student perspective")
```

```python
def parcer(class_obj):
    for i in class_obj:
        i.lec_info()
```

```python
obj1=teacher_lec()
```

```python
obj2=Student_lec()
```

```python
class_obj=[obj1,obj2]
```

```python
parcer(class_obj)
```

```python
class Car:
    def __init__(self,year,make, speed,model):
        self.__year=year
        self.__make=make
        self.__model=model
        self.__speed=speed 
    
    def set_speed(self,speed):
        self.__speed=0 if speed < 0 else speed
    def get_speed(self):
        return self.__speed
```

```python
class Bank:
    def __init__(self,balance):
        self.__balance=balance
    def deposite(self,amount):
        self.__balance=self.__balance+amount
        return "Current Balance",self.__balance
    def withdrwal(self,amount):
        if amount>self.__balance:
            return "Insufficient Balance"
        else:
            self.__balance=self.__balance-amount
            return "remaining amount",self.__balance
        
```

```python
person=Bank(10000)
```

```python
person.deposite(10010)
```

```python
person.withdrwal(111111111)
```

```python
person.withdrwal(20000)
```

```python
person.deposite(110)
```
