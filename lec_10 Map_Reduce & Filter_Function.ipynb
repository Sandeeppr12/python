```python
l=[1,2,3,4,5,6]
```

```python
def fun(l):
    l1=[]
    for i in l:
        l1.append(i**2)
    return l1
```

```python
fun(l)
```

```python
def sq(x):
    return x**2
```

```python
list(map(sq,l))
```

```python
list(map(lambda x:x**2,l))
```
